```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_df.csv')
```

```python
df.info()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
# Data to be predicted
desired_col = df[df['Transported'].isnull()]
desired_col = desired_col.drop(['Transported'], axis = 1)
```

```python
desired_col.head()
```

```python
#Training data
tr_data = df.copy()
tr_data = tr_data.dropna()
```

```python
tr_data.head()
```

```python
X = tr_data.drop('Transported', axis = 1)
y = tr_data['Transported']
```

```python
y = y.astype('int64')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
log_model = LogisticRegression()
```

```python
penalty = ['l1', 'l2', 'elasticnet']
l1_ratio = np.linspace(0,1,20)
C = np.logspace(0,10,20)

param_grid = {'penalty': penalty,
             'l1_ratio': l1_ratio,
             'C': C}
```

```python
grid_model = GridSearchCV(log_model,param_grid=param_grid)
```

```python
grid_model.fit(scaled_X_train,y_train)
```

```python
grid_model.best_params_
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
log_model = LogisticRegression(solver='saga',multi_class="ovr",max_iter=5000, C=1, l1_ratio=0.0, penalty='l1')
```

```python
log_model.fit(scaled_X_train,y_train)
```

```python
y_pred = log_model.predict(scaled_X_test)
```

```python
plot_confusion_matrix(log_model,scaled_X_test,y_test);
```

```python
print(classification_report(y_test,y_pred))
```

```python
log_model.get_params()
```

```python
final_model = LogisticRegression(solver='saga',multi_class="ovr",max_iter=5000, C=1, l1_ratio=0.0, penalty='l1')
```

```python
scaled_X = scaler.fit_transform(X)
sc_desired_col = scaler.fit_transform(desired_col)
```

```python
final_model.fit(scaled_X, y)
```

```python
final_pred = final_model.predict(scaled_X)
```

```python
final_model.coef_[0]
```

```python
coefs = pd.Series(index=X.columns,data=final_model.coef_[0])
coefs = coefs.sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=coefs.index,y=coefs.values)
plt.xticks(rotation=90);
```

```python
print("Logistic Regression Accuracy: "+repr(round(final_model.score(scaled_X, y) * 100, 2)) + "%")
```

```python
result = final_model.predict(sc_desired_col)
```

```python
test = pd.read_csv('test.csv')
```

```python
submission = pd.DataFrame({'PassengerId':test['PassengerId'],'Transported':result})
submission['Transported'] = submission['Transported'].astype(int)
print(submission.shape)
```

```python
submission['Transported'].value_counts()
```

```python
submission['Transported'] = submission['Transported'].astype('bool')
```

```python
filename = 'space_preds.csv'
submission.to_csv(filename,index=False)
print('Saved file: ' + filename)
```
